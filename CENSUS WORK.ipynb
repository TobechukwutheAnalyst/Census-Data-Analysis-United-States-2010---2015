**EXPLORATORY DATA ANALYSIS ON POPULATION DATA FOR COUNTIES & STATES IN THE USA FROM 2010 - 2015**

**WE INTEND TO ANSWER THE QUESTIONS BELOW.**
- 1. Which state has the most counties in it? 
- 2. Looking at the three most populous counties for each state, what are the three most populous states 
- 3. Which city has the most counties in it?
- 4. Which region has the most division in it?

```python
# Importing the required Libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df = pd.DataFrame()
```

```python
# Import and Read file
df = pd.read_csv(r"C:\Users\user\Desktop\Python Project\census.csv")
df
```

**1. Which state has the most counties in it?**

```python
# Group by state name and count the number of unique counties in each state
county_counts = df.groupby('STNAME')['CTYNAME'].nunique()

# Find the state with the maximum number of counties
state_with_most_counties = county_counts.idxmax()

# Print the state with the most counties and its count
print(f"The state with the most counties is {state_with_most_counties} with {county_counts.max()} counties.")
```

```python
# Rename columns
df = df.rename(columns={'CTYNAME': 'County Name', 'STNAME': 'State Name'})
```

```python
# View the names of the columns
print(df.columns)
```

**2. Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Use CENSUS2010POP**

```python
# Group by state and sum the population
state_population = df.groupby('STATE_NAME')['CENSUS2010POP'].sum()

# Sort the states by population in descending order
state_population_sorted = state_population.sort_values(ascending=False)

# Get the top three most populous states
top_three_states = state_population_sorted.head(3)

# Print the result
print("The three most populous states are:")
print(top_three_states)
```

The three most populous are California, Texas and New York

**3. Which city has the most counties in it?**

```python
city_county_counts = df.groupby('CITY').nunique()
city_with_most_counties = city_county_counts.idxmax()
print("City with the most counties:", city_with_most_counties)
```

```python
print(df.columns)
```

```python
# Group by city and count the number of unique counties
city_county_counts = df.groupby('CITY')['COUNTY'].nunique()

# Find the city with the maximum number of unique counties
city_with_most_counties = city_county_counts.idxmax()

# Find the number of counties for the city with the most counties
max_county_count = city_county_counts.max()

print("City with the most counties:", city_with_most_counties)
print("Number of counties in the city:", max_county_count)
```

The Washington county has the largest number of cities in it being 30

**4. Which region has the most division in it?**

```python
# Count occurrences of each unique region
region_counts = df['REGION'].value_counts()

# Find the region with the highest count
region_with_most_data = region_counts.idxmax()

print("REGION with the most data:", region_with_most_data)

```

```python
# Group by region and count the number of unique divisions
region_division_counts = df.groupby('REGION')['DIVISION'].nunique()

# Find the region with the maximum number of unique divisions
region_with_most_divisions = region_division_counts.idxmax()

# Find the number of divisions in the region with the most data
most_divisions_count = region_division_counts.max()

print("Region with the most data:", region_with_most_divisions)
print("Number of divisions in the most data region:", most_divisions_count)
```
